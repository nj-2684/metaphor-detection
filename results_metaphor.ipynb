## Count Visualization:

```python
def count(adf, string, cno):
  # for i in range(len(adf[adf['chapter']==cno])):
    acount_list = adf['count_met_ref'].tolist()
    df = pd.value_counts(np.array(acount_list))
    df = df.reset_index()
    df.columns = ['unique_values', string]
    df.set_index('unique_values')
    return df.sort_values(['unique_values'])

def count_chap(cno):
  df2 = count(mdf, 'Mahatama Gandhi', cno)
  df3 = count(edf, 'Eknath Easwaran', cno)
  df4 = count(sdf, 'Shri Swami Purohit', cno)
  count_df = pd.merge(df2, pd.merge(df3, df4, on='unique_values', how='outer'), on='unique_values', how='outer').sort_values(['unique_values']).fillna(0)
  count_df.set_index('unique_values')
  plotdata = pd.DataFrame({
    "Mahatama Gandhi":count_df['Mahatama Gandhi'].tolist(),
    "Eknath Easwaran":count_df['Eknath Easwaran'].tolist(),
    "Shri Swami Purohit":count_df['Shri Swami Purohit'].tolist()},
    index=count_df['unique_values'].tolist())
  plotdata.plot(kind="bar",figsize=(8,4))
  # plt.title("Count of Metaphors in Chapter {0} for different translations".format(cno))
  plt.xlabel("Number of metaphors")
  plt.ylabel("Number of shlokas")
  plt.savefig('/content/images/tcby_met.png', bbox_inches="tight")
  plt.show()

count_chap("all")
```
